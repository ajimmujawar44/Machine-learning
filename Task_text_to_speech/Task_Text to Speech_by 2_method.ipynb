### Text to Speech conversion using pyttsx3

#### Changing the gender using pyttsx3

```python
#!pip install pyttsx3
```

```python
import pyttsx3
from IPython.display import Audio

text = """
Welcome to Naresh Technology Data Science Program under the guidance of Prakash Senapati.

This training provides practical exposure to improve technical knowledge and coding skills.

We conduct this program for both technical and non-technical learners.

Thank you.
"""

# Initialize engine
engine = pyttsx3.init()

# Speech rate (default is around 200)
engine.setProperty('rate', 150)

# Volume (0.0 to 1.0)
engine.setProperty('volume', 0.8)

# Available voices
voices = engine.getProperty('voices')

# Male voice
engine.setProperty('voice', voices[0].id)

# Female voice (Uncomment if available)
# engine.setProperty('voice', voices[1].id)

# Speak the text
engine.say(text)

# Save as WAV (recommended)
engine.save_to_file(text, 'test_male_voice.wav')

# Run the engine
engine.runAndWait()

print("Audio saved successfully!")

# Display audio player in Jupyter
Audio("test_male_voice.wav")



# ========== we need to install "pip install pyttsx3"=========================
```

```python
import pyttsx3
from IPython.display import Audio

text = """
Welcome to Naresh Technology Data Science Program under the guidance of Prakash Senapati.

This training provides practical exposure to improve technical knowledge and coding skills.

We conduct this program for both technical and non-technical learners.

Thank you.
"""

# Initialize engine
engine = pyttsx3.init()

# Speech rate (default is around 200)
engine.setProperty('rate', 150)

# Volume (0.0 to 1.0)
engine.setProperty('volume', 0.8)

# Available voices
voices = engine.getProperty('voices')

# Male voice
engine.setProperty('voice', voices[0].id)

# Female voice (Uncomment if available)
# engine.setProperty('voice', voices[1].id)

# Speak the text
engine.say(text)

# Save as WAV (recommended)
engine.save_to_file(text, 'test_male_voice.wav')

# Run the engine
engine.runAndWait()

print("Audio saved successfully!")

# Display audio player in Jupyter
Audio("test_male_voice.wav")
```

```python
import pyttsx3

# Initialize engine
engine = pyttsx3.init()

# Set speech rate
engine.setProperty('rate', 150)

# Set volume (0.0 to 1.0)
engine.setProperty('volume', 1.0)

# Get available voices
voices = engine.getProperty('voices')

# Select voice
# 0 = Male, 1 = Female (if available)
engine.setProperty('voice', voices[0].id)

# Text to Speech
text = """
Assalamualaikum to Everyone !!

My name is Azeem Mujawaar.
I am a Data Science student.

I am passionate about Python Programming,
Machine Learning, Data Analysis, and Artificial Intelligence.

My goal is to improve my technical skills,
build real-world projects,
and become a successful Data Scientist.

Thank you for listening,

Allah Hafiz!!
"""

# Speak the text
engine.say(text)

# Save audio file
engine.save_to_file(text, "Azim_Mujawar_Introduction.wav")

# Run engine
engine.runAndWait()

print("Audio file saved successfully!")

# Display audio player in Jupyter
Audio("Azim_Mujawar_Introduction.wav")
```
